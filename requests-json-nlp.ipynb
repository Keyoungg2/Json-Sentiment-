# Question 3. 
Write a function that takes an artist, song, and filename, accesses the lyrics.ovh api to get the song lyrics, and writes the results to the specified filename.  Test this function by getting the lyrics to any four songs of your choice and storing them in different files.

# Question 3 Response

```python
def fetch_lyrics_and_artist(artist, song, filename):
    # Construct for the lyrics API 
    url = f"https://api.lyrics.ovh/v1/{artist}/{song}"

 # Making a GET request to the API
    response = requests.get(url)
    
    # if the request was successful (status code 200)
    if response.status_code == 200:
        # Extracting the lyrics from the response
        data = response.json()
        lyrics = data.get('lyrics')
     # Writing the lyrics to the specified file
        with open(filename, 'w') as file:
            file.write(lyrics)
        print(f"Lyrics for '{song}' by {artist} saved to '{filename}'.")

    # If request is not successful     
    else:
        # response for failuare
        if response.status_code == 404:
            print(f"Failaure fetching lyrics for '{song}' by {artist}.")
            
# Running function to test and see if files are produces with lyrics 
fetch_lyrics_and_artist('Nas', 'One Mic', 'One_Mic_lyrics.txt')
fetch_lyrics_and_artist('Coldplay', 'Viva La Vida', 'Viva_La_Vida_lyrics.txt')
fetch_lyrics_and_artist('Michael Jackson', 'Billie Jean', 'Billie_Jean _lyrics.txt')
fetch_lyrics_and_artist('Inner Circle', 'Bad Boys', 'bad_boys_lyrics.txt')

    

```

 Write a function that takes the name of a file that contains song lyrics, loads the file, performs sentiment analysis, and returns the polarity score. # Question 4 Response

```python
def sentiment_analyze(filename, song_title):
    # Load the file containing song lyrics
    with open(filename, 'r') as file:
        lyrics = file.read()
    
    # Perform sentiment analysis using spaCyTextBlob
    nlp = spacy.load('en_core_web_sm')
    nlp.add_pipe('spacytextblob')
    doc = nlp(lyrics)
    sentiment_score = doc._.polarity
    
    # Print the polarity score with the full song title
    print(f"Polarity score for '{song_title}': {sentiment_score}")
    
    return sentiment_score

# Testing the function with the three files created in question 3
filenames = [
    ('One_Mic_lyrics.txt', 'One Mic'),
    ('Billie_Jean _lyrics.txt', 'Billie Jean'),
    ('bad_boys_lyrics.txt', 'Bad Boys')
]

for file_name, title in filenames:
    polarity_score = sentiment_analyze(file_name, title)
```
